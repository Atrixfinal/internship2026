# Day 4 Assignment – Object-Oriented Programming (OOP)

## Introduction to OOP

Object-Oriented Programming (OOP) is a programming paradigm that organizes code using classes and objects. It helps in creating reusable, scalable, and maintainable programs.

Python supports OOP concepts such as classes, objects, inheritance, polymorphism, encapsulation, and abstraction.

### Advantages of OOP

* Code reusability
* Better organization
* Easier maintenance
* Improved security
* Scalability
* Real-world modeling

---

# 1. Classes and Objects

A class is a blueprint for creating objects.

An object is an instance of a class.

### Example

```python
class Student:
    pass

s1 = Student()
```

### Explanation

* `Student` is a class.
* `s1` is an object of the class.

---

# 2. Constructors (`__init__`)

A constructor is a special method that is automatically called when an object is created.

### Example

```python
class Student:

    def __init__(self, name):
        self.name = name

s1 = Student("Abhinav")
```

### Explanation

* `__init__()` initializes object data.
* It runs automatically during object creation.

---

# 3. Instance Variables and Methods

Instance variables belong to an object.

Instance methods operate on object data.

### Example

```python
class Student:

    def __init__(self, name):
        self.name = name

    def display(self):
        print(self.name)
```

### Explanation

* `name` is an instance variable.
* `display()` is an instance method.

---

# 4. Class Variables

Class variables are shared by all objects of a class.

### Example

```python
class Student:

    school = "ABC School"

    def __init__(self, name):
        self.name = name
```

### Explanation

* `school` is a class variable.
* Shared among all objects.

---

# 5. Encapsulation

Encapsulation means binding data and methods together and restricting direct access to data.

### Example

```python
class BankAccount:

    def __init__(self):
        self.__balance = 1000
```

### Explanation

* `__balance` is private.
* Data is protected from direct access.

---

# 6. Inheritance

Inheritance allows one class to acquire properties and methods of another class.

### Example

```python
class Employee:

    def work(self):
        print("Working")

class Manager(Employee):
    pass
```

### Explanation

* `Manager` inherits from `Employee`.
* Child class can use parent class methods.

---

# 7. Polymorphism

Polymorphism means one interface with multiple forms.

### Example

```python
class Car:

    def start(self):
        print("Car starts")

class Bike:

    def start(self):
        print("Bike starts")
```

### Explanation

* Same method name.
* Different behavior.

---

# 8. Abstraction

Abstraction hides implementation details and shows only essential features.

### Example

```python
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass
```

### Explanation

* Abstract classes contain abstract methods.
* Child classes must implement them.

---

# 9. Method Overriding

Method overriding occurs when a child class provides its own implementation of a parent class method.

### Example

```python
class Animal:

    def sound(self):
        print("Animal Sound")

class Dog(Animal):

    def sound(self):
        print("Bark")
```

### Explanation

* Child class replaces parent method behavior.

---

# 10. super() Function

The `super()` function is used to access parent class methods and constructors.

### Example

```python
class Parent:

    def __init__(self):
        print("Parent Constructor")

class Child(Parent):

    def __init__(self):
        super().__init__()
        print("Child Constructor")
```

### Explanation

* Calls parent class constructor.
* Avoids code duplication.

---

# 11. Access Modifiers

Access modifiers control the visibility of class members.

---

## Public Members

Accessible from anywhere.

### Example

```python
class Student:

    name = "Abhinav"
```

---

## Protected Members

Accessible within the class and its subclasses.

### Example

```python
class Student:

    _name = "Abhinav"
```

---

## Private Members

Accessible only inside the class.

### Example

```python
class Student:

    __name = "Abhinav"
```

---

Object-Oriented Programming helps in creating organized, reusable, and scalable applications. Concepts such as classes, objects, inheritance, polymorphism, encapsulation, and abstraction make programs easier to manage and maintain. These concepts are widely used in real-world software development.

````python
# 3. Attributes and Methods Example

class Student:

    # Constructor
    def __init__(self, name, roll):
        self.name = name
        self.roll = roll

    # Method
    def display(self):
        print("Name:", self.name)
        print("Roll Number:", self.roll)


# Creating object
s1 = Student("Abhinav", 101)

# Calling method
s1.display()
````

````python
# 4. Inheritance Example

class Employee:

    # Parent class method
    def work(self):
        print("Employee is working")


# Child class
class Manager(Employee):

    # Child class method
    def manage(self):
        print("Manager is managing the team")


# Creating object
m1 = Manager()

# Calling inherited method
m1.work()

# Calling child method
m1.manage()
````

````python
# 5. Polymorphism Example

class Car:

    def start(self):
        print("Car starts with key")


class Bike:

    def start(self):
        print("Bike starts with self-start")


# Creating objects
c1 = Car()
b1 = Bike()

# Same method name, different behavior
c1.start()
b1.start()
````

````python
# 6. Encapsulation Example

class BankAccount:

    # Constructor
    def __init__(self, balance):
        self.__balance = balance

    # Method to access private variable
    def get_balance(self):
        return self.__balance


# Creating object
account = BankAccount(5000)

# Accessing balance through method
print("Balance:", account.get_balance())
````

````python
# 7. Abstraction Example

from abc import ABC, abstractmethod


# Abstract Class
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass


# Child Class
class Rectangle(Shape):

    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def area(self):
        return self.length * self.breadth


# Creating object
r1 = Rectangle(10, 5)

# Displaying area
print("Area =", r1.area())
````

````python
# 8. Method Overriding Example

class Animal:

    def sound(self):
        print("Animal makes sound")


class Dog(Animal):

    # Overriding parent method
    def sound(self):
        print("Dog barks")


# Creating object
d1 = Dog()

# Calling overridden method
d1.sound()
````

````python
# 9. Class Variable and Instance Variable Example

class Student:

    # Class Variable
    school = "ABC School"

    # Constructor
    def __init__(self, name):
        self.name = name

    # Method
    def display(self):
        print("Name:", self.name)
        print("School:", Student.school)


# Creating objects
s1 = Student("Abhinav")
s2 = Student("Rahul")

# Displaying details
s1.display()
print()
s2.display()
````

# Practice Programs

````python
# 1. Student Management System

class Student:

    # Constructor
    def __init__(self, name, roll, marks):
        self.name = name
        self.roll = roll
        self.marks = marks

    # Method to calculate grade
    def calculate_grade(self):

        if self.marks >= 90:
            return "A"

        elif self.marks >= 75:
            return "B"

        else:
            return "C"

    # Method to display details
    def display(self):

        print("Name:", self.name)
        print("Roll Number:", self.roll)
        print("Marks:", self.marks)
        print("Grade:", self.calculate_grade())


# Creating object
student1 = Student("Abhinav", 101, 88)

# Displaying details
student1.display()
````

````python
# 2. Bank Account System

class BankAccount:

    # Constructor
    def __init__(self, balance):

        # Private variable
        self.__balance = balance

    # Deposit method
    def deposit(self, amount):

        self.__balance += amount
        print("Deposited:", amount)

    # Withdraw method
    def withdraw(self, amount):

        if amount <= self.__balance:

            self.__balance -= amount
            print("Withdrawn:", amount)

        else:
            print("Insufficient Balance")

    # Check balance method
    def get_balance(self):

        return self.__balance


# Creating object
account = BankAccount(5000)

# Performing operations
account.deposit(2000)
account.withdraw(1000)

# Displaying balance
print("Current Balance:", account.get_balance())
````

````python
# 3. Employee Management System

# Parent Class
class Employee:

    def work(self):

        print("Employee is working")


# Child Class
class Manager(Employee):

    def manage(self):

        print("Manager is managing the team")


# Creating object
manager1 = Manager()

# Calling methods
manager1.work()
manager1.manage()
````

````python
# 4. Vehicle System using Polymorphism

class Vehicle:

    def start(self):

        print("Vehicle starts")


class Car(Vehicle):

    # Overriding method
    def start(self):

        print("Car starts with key")


class Bike(Vehicle):

    # Overriding method
    def start(self):

        print("Bike starts with self-start")


# Creating objects
car = Car()
bike = Bike()

# Calling same method
car.start()
bike.start()
````

````python
# 5. Shape Area Calculator

from abc import ABC, abstractmethod

# Abstract Class
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass


# Child Class
class Rectangle(Shape):

    # Constructor
    def __init__(self, length, breadth):

        self.length = length
        self.breadth = breadth

    # Implementing abstract method
    def area(self):

        return self.length * self.breadth


# Creating object
rectangle = Rectangle(10, 5)

# Displaying area
print("Area of Rectangle =", rectangle.area())
````

````python
# 6. Library Management System

class Book:

    # Constructor
    def __init__(self, title):

        self.title = title
        self.available = True


class Member:

    # Constructor
    def __init__(self, name):

        self.name = name


class Library:

    # Constructor
    def __init__(self):

        self.books = []

    # Add book
    def add_book(self, book):

        self.books.append(book)
        print(book.title, "added successfully")

    # Issue book
    def issue_book(self, title):

        for book in self.books:

            if book.title == title and book.available:

                book.available = False
                print(title, "issued successfully")
                return

        print("Book not available")

    # Return book
    def return_book(self, title):

        for book in self.books:

            if book.title == title:

                book.available = True
                print(title, "returned successfully")
                return

    # Display available books
    def display_books(self):

        print("\nAvailable Books:")

        for book in self.books:

            if book.available:
                print(book.title)


# Creating objects
library = Library()

book1 = Book("Python Programming")
book2 = Book("Data Structures")

# Adding books
library.add_book(book1)
library.add_book(book2)

# Display books
library.display_books()

# Issue book
library.issue_book("Python Programming")

# Display books again
library.display_books()

# Return book
library.return_book("Python Programming")

# Display books again
library.display_books()
````

# OOP Interview Questions and Answers
---

## 1. What is a Class?

**Answer:**

A class is a blueprint or template used to create objects. It defines the attributes (data) and methods (functions) that objects will have.

**Example:**

```python
class Student:
    pass
```

---

## 2. What is an Object?

**Answer:**

An object is an instance of a class. It represents a real-world entity and can access the attributes and methods defined in the class.

**Example:**

```python
s1 = Student()
```

---

## 3. Difference Between Class and Object?

| Class                          | Object                          |
| ------------------------------ | ------------------------------- |
| Blueprint or template          | Instance of a class             |
| Logical entity                 | Physical entity                 |
| Defines properties and methods | Uses properties and methods     |
| Created once                   | Multiple objects can be created |

**Example:**

```python
class Student:
    pass

s1 = Student()
s2 = Student()
```

---

## 4. What is Inheritance?

**Answer:**

Inheritance is an OOP feature that allows one class to acquire properties and methods of another class. It promotes code reusability.

**Example:**

```python
class Employee:
    pass

class Manager(Employee):
    pass
```

---

## 5. What is Polymorphism?

**Answer:**

Polymorphism means "many forms." It allows the same method name to behave differently in different classes.

**Example:**

```python
class Car:
    def start(self):
        print("Car Starts")

class Bike:
    def start(self):
        print("Bike Starts")
```

---

## 6. What is Encapsulation?

**Answer:**

Encapsulation is the process of binding data and methods together and restricting direct access to data using private variables.

**Example:**

```python
class Bank:

    def __init__(self):
        self.__balance = 5000
```

**Benefits:**

* Data security
* Better control over data
* Prevents unauthorized access

---

## 7. What is Abstraction?

**Answer:**

Abstraction hides implementation details and shows only the essential features to the user.

**Example:**

```python
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass
```

**Benefits:**

* Reduces complexity
* Improves maintainability
* Enhances security

---

## 8. Difference Between Method Overriding and Method Overloading?

| Method Overriding                          | Method Overloading                    |
| ------------------------------------------ | ------------------------------------- |
| Same method in parent and child class      | Same method with different parameters |
| Achieved using inheritance                 | Not directly supported in Python      |
| Runtime polymorphism                       | Compile-time concept                  |
| Child class changes parent method behavior | Multiple versions of same method      |

**Method Overriding Example:**

```python
class Animal:

    def sound(self):
        print("Animal Sound")

class Dog(Animal):

    def sound(self):
        print("Dog Barks")
```

---

## 9. What is the Purpose of `super()`?

**Answer:**

The `super()` function is used to access parent class methods and constructors from the child class.

**Example:**

```python
class Parent:

    def __init__(self):
        print("Parent Constructor")

class Child(Parent):

    def __init__(self):
        super().__init__()
        print("Child Constructor")
```

**Advantages:**

* Reuses parent class code
* Avoids duplication
* Makes inheritance easier

---
