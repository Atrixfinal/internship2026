# Introduction to Functions in Python

A function is a block of reusable code used to perform a specific task. Functions help in reducing code repetition and make programs more organized, readable, and maintainable.

Functions allow programmers to divide a large problem into smaller manageable parts.

---

# Advantages of Functions

* Code reusability
* Reduces repetition
* Improves readability
* Easier debugging
* Better program organization
* Simplifies maintenance

---

# Function Definition and Scope

A function is defined using the `def` keyword.

## Syntax

```python
def function_name():
    # code block
```

## Example

```python
def greet():
    print("Hello")
```

### Explanation

* `def` is used to define a function.
* `greet` is the function name.
* The code inside the function executes when the function is called.

---

# Function Calling

A function runs only when it is called.

## Example

```python
def greet():
    print("Hello")

greet()
```

### Output

```python
Hello
```

---

# Parameters and Arguments

Parameters are variables used in function definition, while arguments are actual values passed during function calling.

## Example

```python
def add(a, b):
    print(a + b)

add(5, 3)
```

### Explanation

* `a` and `b` are parameters.
* `5` and `3` are arguments.

### Output

```python
8
```

---

# Types of Arguments

## 1. Positional Arguments

Values are assigned according to their position.

```python
def student(name, age):
    print(name, age)

student("Rahul", 20)
```

---

## 2. Keyword Arguments

Arguments are passed using parameter names.

```python
def student(name, age):
    print(name, age)

student(age=20, name="Rahul")
```

---

## 3. Default Arguments

A default value is assigned to a parameter.

```python
def course(name, subject="Python"):
    print(name, subject)

course("Abhinav")
```

### Output

```python
Abhinav Python
```

---

## 4. Variable-Length Arguments (`*args`)

Used when the number of arguments is unknown.

```python
def numbers(*args):
    print(args)

numbers(1, 2, 3, 4)
```

### Output

```python
(1, 2, 3, 4)
```

# Return Values and Data Types


* `return` stores the output of the function.

# Recursion Concepts

Recursion is a process where a function calls itself repeatedly until a stopping condition is met.

## Example: Factorial Using Recursion

```python
def factorial(n):

    if n == 1:
        return 1

    else:
        return n * factorial(n - 1)

print(factorial(5))
```

### Output

```python
120
```

### Important Terms

## Base Condition

The condition that stops recursion.

## Recursive Call

The function calling itself.

---

# Variable Scope

Scope defines where a variable can be accessed in a program.

---

# Types of Scope

## 1. Local Scope

Variables declared inside a function are local variables.

```python
def demo():
    x = 10
    print(x)

demo()
```

### Explanation

* `x` exists only inside the function.

---

## 2. Global Scope

Variables declared outside functions are global variables.

```python
x = 100

def demo():
    print(x)

demo()
```

### Explanation

* Global variables can be accessed anywhere in the program.

---

## 3. Nonlocal Scope

Used in nested functions to modify variables of the outer function.

## Example

```python
def outer():

    x = "Python"

    def inner():

        nonlocal x
        x = "Programming"

    inner()

    print(x)

outer()
```

### Output

```python
Programming
```

# Code Reusability

Code reusability means using the same code multiple times instead of rewriting it.

Functions improve reusability by:

* Breaking programs into smaller parts
* Reusing functions whenever needed

# Helper Functions

Helper functions are small functions created to support larger functions.

## Example

```python

    print(square(num))
```

```python

```

# Built-in Functions Used with Functions

---

## map()

Applies a function to every item in a sequence.

```python
numbers = [1, 2, 3]

result = list(map(lambda x: x * 2, numbers))

print(result)
```

### Output

```python
[2, 4, 6]
```

---

## filter()

Filters elements based on a condition.

```python
numbers = [1, 2, 3, 4]

result = list(filter(lambda x: x % 2 == 0, numbers))

print(result)
```

### Output

```python
[2, 4]
```

---

## reduce()

Reduces all elements into a single value.

```python
from functools import reduce

numbers = [1, 2, 3, 4]

result = reduce(lambda x, y: x + y, numbers)

print(result)
```

### Output

```python
10
```

````python
## 6. Recursive Function Example - Factorial

# Function to calculate factorial using recursion
def factorial(n):

    # Base condition
    # Factorial of 1 is always 1
    if n == 1:
        return 1

    # Recursive call
    else:
        return n * factorial(n - 1)

# Calling the function
print("Factorial =", factorial(5))
````

````python
## 8.  Helper Functions for Code Reusability

# Helper function to calculate square
def calculate_square(n):

    # Returning square value
    return n * n

# Another function using helper function
def display(number):

    # Calling helper function
    result = calculate_square(number)

    # Displaying result
    print("Square =", result)

# Calling display function
display(6)
````

````python
## 1.  BMI Calculator Program

# Function to calculate BMI
def calculate_bmi(weight, height):

    # BMI Formula
    bmi = weight / (height ** 2)

    # Returning BMI value
    return bmi


# Function to display BMI category
def bmi_category(bmi):

    # Checking BMI ranges
    if bmi < 18.5:
        return "Underweight"

    elif bmi < 25:
        return "Normal Weight"

    elif bmi < 30:
        return "Overweight"

    else:
        return "Obese"


# Taking input from user
weight = float(input("Enter weight in kg: "))
height = float(input("Enter height in meters: "))

# Calling BMI function
bmi = calculate_bmi(weight, height)

# Displaying BMI value
print("BMI Value =", round(bmi, 2))

# Displaying BMI category
print("Category =", bmi_category(bmi))
````

````python
## 2.  EMI Calculator Program

# Importing math module
import math

# Function to calculate EMI
def calculate_emi(principal, rate, time):

    # Converting annual interest rate into monthly interest rate
    monthly_rate = rate / (12 * 100)

    # Converting years into months
    months = time * 12

    # EMI Formula
    emi = (principal * monthly_rate * math.pow(1 + monthly_rate, months)) / \
          (math.pow(1 + monthly_rate, months) - 1)

    # Returning EMI value
    return emi


# Taking input from user
principal = float(input("Enter Loan Amount: "))
rate = float(input("Enter Annual Interest Rate: "))
time = int(input("Enter Loan Tenure (Years): "))

# Calling EMI function
emi = calculate_emi(principal, rate, time)

# Displaying EMI
print("Monthly EMI =", round(emi, 2))
````

````python
## 3.  Recursive Factorial Program

# Function to calculate factorial using recursion
def factorial(n):

    # Base condition
    if n == 0 or n == 1:
        return 1

    # Recursive call
    else:
        return n * factorial(n - 1)


# Taking input from user
number = int(input("Enter a number: "))

# Calling factorial function
result = factorial(number)

# Displaying factorial
print("Factorial =", result)
````

````python
## 4.  Fibonacci Series Generator
         ## i. Fibonacci Series using Function

# Function to generate Fibonacci series
def fibonacci(n):

    # First two numbers
    a = 0
    b = 1

    print("Fibonacci Series:")

    # Loop for generating series
    for i in range(n):

        print(a, end=" ")

        # Updating values
        a, b = b, a + b


# Taking input from user
terms = int(input("Enter number of terms: "))

# Calling function
fibonacci(terms)
````

````python

   ## ii. Fibonacci Series using Recursion

# Recursive function
def fibonacci(n):

    # Base conditions
    if n <= 1:
        return n

    # Recursive calls
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)


# Taking input from user
terms = int(input("Enter number of terms: "))

print("Fibonacci Series:")

# Loop for displaying series
for i in range(terms):
    print(fibonacci(i), end=" ")
````

````python
## 5. Prime Number Checker Program

# Function to check prime number
def is_prime(number):

    # Numbers less than or equal to 1 are not prime
    if number <= 1:
        return False

    # Checking divisibility
    for i in range(2, number):

        # If divisible, not prime
        if number % i == 0:
            return False

    # Otherwise prime
    return True


# Taking input from user
num = int(input("Enter a number: "))

# Calling function
if is_prime(num):
    print(num, "is a Prime Number")

else:
    print(num, "is not a Prime Number")
````

````python
## 6.  List Operations Function
       ## i. Using map() Function

# List of numbers
numbers = [1, 2, 3, 4, 5]

# Using map() to calculate square of each number
squares = list(map(lambda x: x * x, numbers))

# Displaying result
print("Squared List:", squares)
````

````python

       ## ii. Using filter() Function

# List of numbers
numbers = [1, 2, 3, 4, 5, 6]

# Filtering even numbers
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

# Displaying result
print("Even Numbers:", even_numbers)
````

````python

       ## iii. Using reduce() Function

# Importing reduce from functools
from functools import reduce

# List of numbers
numbers = [1, 2, 3, 4, 5]

# Using reduce() to calculate sum of all numbers
total = reduce(lambda x, y: x + y, numbers)

# Displaying result
print("Sum of Numbers =", total)
````
